```python
# imports
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# set dir to raw data files
filedir = "C:\\Users\\Elly\\Uni\\Masterarbeit\\data\\raw\\"

# load all needed data
enrol = dt = pd.read_csv(filedir + "\\ENROL.csv", sep=',')
```

```python
# plot overview of patients - which age and sex to they have?

axes = enrol.boxplot(by="sex", column="age", fontsize=12, grid=True)

plt.title("Overview of patients (n=" + str(len(enrol["sex"])) + ")", fontsize = 18)
plt.suptitle('') # that's what you're after

plt.xlabel("sex", fontsize = 16)
axes.set_xticklabels(["Female (n=" + str(len(enrol.sex.loc[enrol.sex == "Female"])) + ")", 
                      "Male (n=" + str(len(enrol.sex.loc[enrol.sex == "Male"])) + ")"])
plt.ylabel("age", fontsize = 16)

plt.show()

# How are the numbers for cip1 and cip2&cip3 patients (screening vs screening 2.0)?
axes = enrol.groupby(["sex","visit"]).size().unstack(1).plot.bar(rot=0,fontsize=12, grid=True)

plt.title("Overview of patients (n=" + str(len(enrol["sex"])) + ")", fontsize = 18)
plt.suptitle('') # that's what you're after

plt.xlabel("sex", fontsize = 16)
axes.set_xticklabels(["Female (n=" + str(len(enrol.sex.loc[enrol.sex == "Female"])) + ")", 
                      "Male (n=" + str(len(enrol.sex.loc[enrol.sex == "Male"])) + ")"])
plt.ylabel("number of patients", fontsize = 16)
plt.legend(loc=2, prop={'size': 12})

plt.show()
```
